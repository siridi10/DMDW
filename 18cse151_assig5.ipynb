```
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sns;
```

```
#Custom dataset 
path="https://raw.githubusercontent.com/Anasuya-Sahoo/DMDW-Lab/main/student-mat.csv"
import pandas as pd
data=pd.read_csv(path)
data.shape
```

```
data.head()
```

```
#BINARY ATTRIBUTES
```

```
d=data[['schoolsup','famsup','paid','activities','nursery','higher','internet','romantic']]
d.head()
```

```
d=d.replace('no',0)
d=d.replace('yes',1)
```

```
#BINARY ATTRIBUTES
```

```
n=np.array(d[['schoolsup','famsup']])
n=n.reshape(-1,2)
n.shape
```

```
m=np.array(d[['internet','romantic']])
m=m.reshape(-1,2)
m.shape
```

```
from scipy.spatial import distance
dist_matrix=distance.cdist(n,m)
dist_matrix.shape
print(dist_matrix)
```

```
sns.heatmap(dist_matrix)
pt.show
```

```
#Nominal Attribute
```

```
nominal=data[['Mjob','Fjob','reason','guardian']]
nominal=nominal.replace('at_home','home')
nominal=(nominal.astype('category'))
```

```
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
nominal['Mjob']=lb.fit_transform(nominal['Mjob'])
nominal['Fjob']=lb.fit_transform(nominal['Fjob'])
nominal['reason']=lb.fit_transform(nominal['reason'])
nominal['guardian']=lb.fit_transform(nominal['guardian'])
nominal.head()
```

```
nom1=np.array(nominal)
nom1=nom1.reshape(-1,2)
```

```
nom2=np.array(nominal)
nom2=nom2.reshape(-1,2)
```

```
from scipy.spatial import distance
dist_matrix=distance.cdist(nom1,nom2)
```

```
sns.heatmap(dist_matrix)
dist_matrix.shape
```
